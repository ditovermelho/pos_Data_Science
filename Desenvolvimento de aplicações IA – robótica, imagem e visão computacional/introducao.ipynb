```python
import cv2
import matplotlib.pyplot as plt
```

```python
imagem = cv2.imread('./img/Screenshot_20240421_164523_Opera.jpg')
```

```python
imgColor = cv2.cvtColor(imagem, cv2.COLOR_RGB2BGR)
imgGray = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(imgColor)
plt.title('imagem normal')

plt.subplot(1, 2, 2)
plt.imshow(imgGray)
plt.title('imagem cinza')

plt.subplots_adjust(wspace=0.3, hspace=0.3)
plt.tight_layout()
plt.show()
```

```python
imgRavel = imagem.ravel()
imgRavelpb = imgGray.ravel()

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(imgRavel)
plt.title('imagem normal')

plt.subplot(1, 2, 2)
plt.hist(imgRavelpb)
plt.title('imagem cinza')

plt.subplots_adjust(wspace=0.3, hspace=0.3)
plt.tight_layout()
plt.show()
```
